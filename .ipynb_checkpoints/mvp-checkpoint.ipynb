```python
mvp = pd.read_csv('mvps.csv', encoding = 'latin-1', delimiter= ';')

nickname = pd.read_csv('nicknames.csv', encoding = 'latin-1', delimiter= ';')

players = pd.read_csv('players.csv', encoding = 'latin-1', delimiter= ';')

teams = pd.read_csv('teams.csv', encoding = 'latin-1', delimiter= ';')
teams.head()
```

```python
players["Player"] = players["Player"].str.replace("*","", regex=False)
players = players[~players["Age"].str.contains("Age")]
players.to_csv('clean_players.csv', index=False)
print(players.columns)
players
```

```python
new_row = pd.Series(data={'Abbreviation':'TOT', 'Name':'Multiple Teams Total'}, name='40')
#append row to the dataframe
nickname = nickname.append(new_row, ignore_index=False)
```

```python
teams.info()
```

```python
teams = teams[~teams["W"].str.contains("Division")]
teams.head(10)
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
teams.head(8)
```

```python
teams["GB"] = teams["GB"].str.replace("","0").astype(float)
teams.head(10)
```

```python
teams = teams.rename(columns={'Team': 'Name'})
```

```python
teams = pd.merge(teams, nickname, on='Name', how='left')
```

```python
print(mvp.columns)
mvp = mvp.rename(columns={'Pts Won': 'Pts_Won', 'Pts Max': 'Pts_Max'})
mvp.head(10)
```

```python
nickname.to_sql('nickname', connection, if_exists='append', chunksize= 100, index=False)
```

```python
teams.to_sql('teams', connection, if_exists='append', chunksize= 100, index=False)
```

```python
mvp.to_sql('mvp', connection, if_exists='append', chunksize= 100, index=False)
```

```python
players.to_sql('players', connection, if_exists='append', chunksize= 100, index=False)
```
